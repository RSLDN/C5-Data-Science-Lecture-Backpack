```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age_df = pd.read_excel("Age.xlsx")
```

```python
age_df.head()
```

# EDA

```python
age_df.info()
```

```python
# pairplot of the data
sns.pairplot(age_df)
plt.show()
```

```python
# Plot distribution of age variable
age_df['age'].plot(kind='hist', bins=30, color='c', edgecolor='black', linewidth=1.2)
```

```python
# heatmap of the data
sns.heatmap(age_df.corr(), annot=True)
plt.show()
```

```python
# Select only 'Top 10' and 'age' columns
selective_age_df = age_df[['Top 10', 'age']]
```

```python
selective_age_df.head()
```

```python
# pairplot of the data
sns.pairplot(selective_age_df)
plt.show()
```

```python
# heatmap of the data
sns.heatmap(selective_age_df.corr(), annot=True)
plt.show()
```

```python
selective_age_df['Top 10'].plot(kind='hist', bins=30, color='c', edgecolor='black', linewidth=1.2)
```
